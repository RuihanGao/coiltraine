```python
float_data = np.random.randn(10, 3)
print(float_data.shape)
meta_data = {'speed': 0, 'brake': 1, 'throttle': 2}
```

```python
keys = lambda_splitter(float_data, meta_data, [
    lambda x,y: np.where(                
        np.logical_and(x[:, 0]> 0., x[:, 1]>0))[0], # speed and brake > 0
    lambda x,y: np.where(
        x[:, 2] > 0)[0]
    ])
```

```python
weights = np.asarray([.9, .1])
```

```python
Sampler = PreSplittedSampler(keys, 10, weights)
```

```python
cc = 0
out = []
for i in Sampler:
    print(i)
    out.append(i)
    cc += 1
    if cc == 1000:
        break
```
